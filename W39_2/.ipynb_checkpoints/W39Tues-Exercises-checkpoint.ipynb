> **Exercise**: Imagine that we have a car that can fit 5 passengers, and we want to know, if the car can fit more passengers. Using 'if' and 'else', make a piece of code that prints out "The car is full" if passengers is 5 or more, and "The car is not full yet" if passengers are less than 5.

```python
# First if statement
# Input: 
# Output:

# Number of places in a car
car = 5

# Number of passangers
passangers = 4

# Creating if statement
if passangers >= car:
    print("The car is full")
else:
    print("The car is not full yet")
```

In the exercise above you created a dictionary of sodas people and their twitter handles. Later, when you start using the Twitter API, you might want to check if a person has a twitter handle before trying to search for them on Twitter. 

>**Exercise:** Using the `sodas_twitter_dict`, write an if statement that checks if a person has a Twitter handle. If they do, print out the handle. If they don't, print out 'No Twitter handle available.' 

>*Hint: First create a variable `name`, which is equal to the name of the person you want to search. You can then change the value of the variable to test your if-statement*

>*Hint 2: You'll need to use the pandas `pd.isna(value)` to check the value of the twitter handle, which returns True if the value is equivalent to nan (i.e. nothing) and False otherwise.*


```python
# Going thourgh the ealier created dic with if clauses

name = "Sune Lehmann"

if sodas_twitter_dic[name] == pd.isna(name):
    print("No Twitter handle available")
else:
    print(sodas_twitter_dic[name])
    

# Alternative with double negation

name_a = "Nicklas Johansen"

if (not pd.isna(sodas_twitter_dic[name_a])): #pd.isna shows if there is nothing there
    print(sodas_twitter_dic[name_a])
else: 
    print("No Twitter handle available")
```

>**Exercise**: Create a program which produces statements about numbers that you give it, with the following conditions:
- if the number is over 100 print "Phew, that's a big number."
- if the number is even, print "This one is even."
- if the number is even and over 100 print "Stop. I can't even."
- if the number doesn't satisfy any of these conditions, print the number.

>*Recall: Multiple conditions can be inserted with elif statement(s). And remember to think through your solution first before jumping to write your code!*

```python
# Check the program with a few values to see if it gives the results you expect.

number = 300

# Check if the nubers is even or over 100
if number > 100 and number % 2 == 0:
    print("Stop I cant even.")

# if the above if is false, then the number is either less than 100
elif number > 100:
    print ("Pew, thats a big number")
elif number % 2 == 0:
    print ("This one is even.")

# if nothing else applies
else: 
    print(number)
```
